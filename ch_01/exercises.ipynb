## Exercise 4: Generate the data by running this cell
This will give you a list of numbers to work with in the remaining exercises.

```python
import random

random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

## Exercise 5: Calculating statistics and verifying
### mean

```python
mean = sum(salaries)/len(salaries)
```

### median

```python
salaries.sort()
salaries[50]
```

### mode

```python
from collections import Counter

Counter(salaries).most_common(1)[0][0]
```

### sample variance
Remember to use Bessel's correction.

```python
mean = sum(salaries)/len(salaries)

sum(map( lambda x: (x - mean)**2, salaries)) / len(salaries)-1
```

### sample standard deviation
Remember to use Bessel's correction.

```python
from math import sqrt

sqrt(69957413899.0)
```

## Exercise 6: Calculating more statistics
### range

```python
max(salaries) - min(salaries)
```

### coefficient of variation
Make sure to use the sample standard deviation.

```python
264494.6386961369 / 803000.0
```

### interquartile range

```python
salaries.sort()
salaries[75] - salaries[25]
```

### quartile coefficent of dispersion

```python
salaries.sort()

(salaries[75] - salaries[25]) / (salaries[75] + salaries[25])
```

## Exercise 7: Scaling data
### min-max scaling

```python
list(map( lambda x: (x - min(salaries)) / (max(salaries) - min(salaries)), salaries))
```

### standardizing

```python
list(map( lambda x: (x - 585690.0) / 264494.6386961369, salaries))
```

## Exercise 8: Calculating covariance and correlation
### covariance

```python
X = list(map( lambda x: (x - min(salaries)) / (max(salaries) - min(salaries)), salaries))
Y = list(map( lambda x: (x - 585690.0) / 264494.6386961369, salaries))

import numpy as np

# Sample data: each row is a variable, each column is an observation (default)
#x = np.array([X,Y])

# Calculate the covariance matrix
cov_matrix = np.cov(X,Y)
print(cov_matrix)
```
